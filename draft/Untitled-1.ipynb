```python
import pandas as pd
```

```python
categories = pd.read_json('elaf_categories.json_cleaned.json')
categories['catSeq'] = range(1000, len(categories) + 1000)
categories
```

```python
with open('elaf_categories_with_sequential.json', 'w', encoding='utf-8') as f:
    categories.to_json(f, force_ascii=False, orient='records', indent=4)
```

```python
products = pd.read_json('elaf_products.json_cleaned.json')
products['proSeq'] = range(1000000, len(products) + 1000000)
products
```

```python
newProducts = products.merge(
    categories[['id', 'catSeq']],
    left_on='categoryId',
    right_on='id',
    how='left',
    suffixes=('_prod', '_cat')
).drop(columns=['id_cat']).rename(columns={'id_prod': 'id'})

newProducts['combinedSeq'] = (
    newProducts['catSeq'].astype(str) + newProducts['proSeq'].astype(str)
)

newProducts = newProducts.drop(columns=['proSeq', 'catSeq'])

with open('elaf_products_with_combinedSeq.json', 'w', encoding='utf-8') as f:
    newProducts.to_json(f, force_ascii=False, orient='records', indent=4)
```
