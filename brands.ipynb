```python
import pandas as pd
import json

df_sheets = pd.read_excel("./ElafProducts.xlsx", sheet_name=None)

def clean_sheet(df):
    df = df.dropna(axis=1, how="all")
    if df.columns.str.contains("Unnamed").all() or all([str(c) == "nan" for c in df.columns]):
        df.columns = df.iloc[0]
        df = df[1:]
    df.columns = df.columns.astype(str)
    df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
    return df

sheets = {name: clean_sheet(df) for name, df in df_sheets.items()}

unique_brands = set()

# Loop through all sheets
for sheet_name, df in sheets.items():
    # Try to get the column containing brand names
    for col in df.columns:
        if "العلامه التجاريه" in col or "العلامة التجارية" in col:
            # Add all non-null brand names to the set
            brands = df[col].dropna().astype(str).str.strip()
            unique_brands.update(brands)

# Convert set to a sorted list
unique_brands_list = sorted(unique_brands)

# Save to JSON
with open("brands.json", "w", encoding="utf-8") as f:
    json.dump(unique_brands_list, f, ensure_ascii=False, indent=2)

print(f"Saved {len(unique_brands_list)} unique brands to brands.json")
```
