```python
import pandas as pd

fileName = 'elaf_products.json'
df = pd.read_json(fileName)

original_count = len(df)

df_cleaned = df.drop_duplicates(subset=['id'], keep='first')

cleaned_count = len(df_cleaned)

deleted_rows = original_count - cleaned_count
print("Deleted rows:", deleted_rows)

output_file = f"{fileName}_cleaned.json"
with open(output_file, 'w', encoding='utf-8') as f:
    df_cleaned.to_json(f, orient='records', force_ascii=False, indent=4)
```
